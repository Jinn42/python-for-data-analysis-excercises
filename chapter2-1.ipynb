```python
# Counting Time Zones (tz) with pandas
```

```python
#Creating a DataFrame from the original set of records
from pandas import DataFrame, Series
import pandas as pd
frame = DataFrame(records)
frame.info()
frame['tz'][:10]
```

```python
frame
```

```python
#the first 10 most frequent time zones
tz_counts = frame['tz'].value_counts()
tz_counts[:10]
```

```python
# make a plot of this data
```

```python
#The fillna function can replace missing (NA) values
#and unknown (empty strings) values can be replaced by boolean array indexing
clean_tz = frame['tz'].fillna('Missing')
clean_tz[clean_tz == ''] = 'Unknown'
tz_counts = clean_tz.value_counts()
tz_counts[:10]
```

```python
#horizontal bar plot
tz_counts[:10].plot(kind='barh', rot=0)
```

```python
#split off the first token in the string and make another summary of the user behavior:
s=frame.a.dropna()

frame.a.dropna()
```

```python
results = Series([x.split()[0] for x in frame.a.dropna()])
results[:5]
```

```python
results.value_counts()[:8]
```

```python
# decompose the top time zones into Windows and non-Windows users
import numpy as np
cframe = frame[frame.a.notnull()]
#We want to then compute a value whether each row is Windows or not:
operating_system = np.where(cframe['a'].str.contains('Windows'), 'Windows', 'Not Windows')
operating_system[:5]
```

```python
# group the data by its time zone column 
by_tz_os = cframe.groupby(['tz', operating_system])
#size() is similar with value_counts(). The result is reshaped into a table with unstack().
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

```python
#construct an indirect index array from the row counts in agg_counts
indexer = agg_counts.sum(1).argsort()

indexer[:10]
```

```python
# use take to select the rows in that order, then slice off the last 10 rows
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
# plotted in a bar plot
count_subset.plot(kind='barh', stacked=True)
```

```python
#to see the relative percentage of Windows users in the smaller groups
# the rows can easily be normalized to sum to 1
normed_subset = count_subset.div(count_subset.sum(1), axis=0)
normed_subset.plot(kind='barh', stacked=True)
```
